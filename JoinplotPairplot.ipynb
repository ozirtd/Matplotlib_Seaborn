```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist',
              marginal_ticks=True, #muestra una pequeña tabla para la gráfica exterior
              marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')) #marginal_kws se le pueden agregar diferentes parámetros
```

```python
sns.pairplot(data=tips)#cruza variables NUMERICAS
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
