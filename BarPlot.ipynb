```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population =[1500,800,900,1000,300]
```

```python
plt.style.use('seaborn-darkgrid')
plt.bar(country,population,width=0.5,color=['#5c58b6','#b457ce','#5994ce','#3a4e93'])
plt.xticks(np.arange(5),('India', 'Japan', 'Mexico', 'Colombia', 'Germany'), rotation=45)
plt.title('Population by country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
countries = ['Perú', 'Argentina', 'Colombia', 'Mexico', 'Brasil']
population = [25.199748, 36.24159, 39.234062, 98.821456, 170.179643]
colors = ['firebrick', 'skyblue', 'gold', 'green', 'limegreen']

plt.style.use("seaborn")
fig, ax = plt.subplots()
bars = ax.barh(countries, population, color=colors, label="Cant. habitantes")

ax.bar_label(bars)
ax.set_xlabel("Cant. de habitantes ($millones$)")
ax.set_title("Países con mas población de Latam")

#plt.show()
plt.savefig(os.path.join('grafica_paises.png'), dpi=600, format='png', bbox_inches='tight')
```
